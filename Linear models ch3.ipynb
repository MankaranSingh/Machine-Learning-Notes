```python
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
%matplotlib notebook
from sklearn.datasets import fetch_mldata
import sklearn
import matplotlib 
```

```python
X = 2*np.random.randn(100,1)
```

```python
X
```

```python
y = 4 + 3*X + np.random.randn(100, 1)  # np.random.randn(100, 1) is known as 'Gaussian Noise'
```

```python
y
```

```python
plt.plot(X,y,'o', alpha = 0.8)
```

```python
X_b = np.c_[np.ones((100, 1)), X]
```

```python
X_b
```

```python
np.ones((100, 1))
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
np.array([[0], [2]])
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
```

```python
X_new_b
```

```python
y_predict = X_new_b.dot(theta_best)
```

```python
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "bo", alpha = 0.8)
plt.axis([0, 2, 0, 15])
plt.show()
```
